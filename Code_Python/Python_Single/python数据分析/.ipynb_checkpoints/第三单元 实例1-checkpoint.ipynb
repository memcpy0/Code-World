## 一、图像的数组表示

图像一般使用RGB色彩模式，即每个像素点的颜色由红、绿、蓝组成。

通过RGB三个颜色通道的变化和叠加得到各种颜色，其中：R G B 的取值范围均为 0-255。

RGB形成的颜色包括了人类视力所能感知的所有颜色。

### PIL库  这是一个具有强大图像处理能力的第三方库

from PIL import Image
 
Image是PIL库中一个代表图像的类


图像是一个由像素组成的二维矩阵，每个元素是一个RGB值（R，G，B）



```python
from PIL import Image

import numpy as np

im = np.array(Image.open("C:\\Users\\dell\\Desktop\\视频音频PPT图片PSD素材\\新视野素材库\\image-6.jpeg"))  #以后文件路径全部改为\\

print(im.ndim,im.shape,im.size,im.dtype)  

#轴（维度的数量），图像（是一个三维数组）的维度（高度，宽度和像素RGB值），图像的元素（像素点）个数，……

#其中高度和宽度即为像素点几行几列！
```

## 二、图像的变换

读入图像后，获得像素RGB值，修改后保存为新的文件

```python
from PIL import Image
import numpy as np
a = np.array(Image.open("horse.jpg"))
print(a.shape,a.size,a.dtype)
```

```python
b = [255,255,255] -a
im = Image.fromarray(b.astype('uint8'))
im.save("C:\\Users\\dell\\demo\\python数据分析\\newhorse.jpg")
```

```python
a = np.array(Image.open("horse.jpg").convert('L'))
b = 255-a
im = Image.fromarray(b.astype('uint8'))
im.save("C:\\Users\\dell\\demo\\python数据分析\\newhorse1.jpg")
```

```python
c = (100/255) *a +150 #区间变换
im = Image.fromarray(c.astype('uint8'))
im.save("C:\\Users\\dell\\demo\\python数据分析\\newhorse2.jpg")
```

```python
d = 255 * (a/255)**2 # 像素平方
im = Image.fromarray(d.astype('uint8'))
im.save("C:\\Users\\dell\\demo\\python数据分析\\newhorse3.jpg")
```

## 四、“图像的手绘效果”实例编写和解读

```python
from PIL import Image
import numpy as np

a = np.asarray(Image.open("C:\\Users\\dell\\Desktop\\视频音频PPT图片PSD素材\\新视野素材库\\地狱星.jpeg").convert('L')).astype('float')
```

```python
# 梯度的重构
# 利用像素之间的梯度值和虚拟深度值对图像进行重构
# 根据灰度变化来模拟人类视觉的远近程度

depth = 10.  # 预设深度值为10 取值范围0‐100 
grad = np.gradient(a)

grad_x,grad_y = grad  # 提取x和y方向的梯度值 

grad_x = grad_x*depth/100. # 根据深度调整x和y方向的梯度值
grad_y = grad_y*depth/100.

# 梯度归一化
# 构造x和y轴梯度的三维归一化单位坐标系
A = np.sqrt(grad_x**2 + grad_y**2 + 1.)  
uni_x = grad_x/A
uni_y = grad_y/A
uni_z = 1./A

vec_e1 = np.pi/2.2
vec_az = np.pi/4
```

```python
# 光源效果
# dx, dy, dz是光源对x/y/z三方向的影响程度
dx = np.cos(vec_e1)*np.cos(vec_az) 
# np.cos(vec_el)为单位光线在地平面上的投影长度
dy = np.cos(vec_e1)*np.sin(vec_az)  
dz = np.sin(vec_e1)

# 梯度与光源相互作用，将梯度转化为灰度
b = 255*(dx*uni_x + dy*uni_y + dz*uni_z)  

b = b.clip(0,255) # 为避免数据越界，将生成的灰度值裁剪至0‐255区间

# 图像生成
im = Image.fromarray(b.astype('uint8'))
im.save("C:\\Users\\dell\\Desktop\\视频音频PPT图片PSD素材\\新视野素材库\\地狱星手绘.jpeg")
```

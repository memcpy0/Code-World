```python
Filename = "Yes.txt"
with open(Filename,'r',encoding="utf-8") as f:
   mytext = f.read()
   print(mytext)
    
```

```python
from wordcloud import WordCloud
from PIL import Image
#from os import path
import matplotlib.pyplot as plt
import numpy as np

mask0 = np.array(Image.open("C:\\Users\\dell\\demo\\255fk.jpg_595.jpg"))
wc = WordCloud(background_color='white',mask=mask0)
wordcloud = wc.generate(mytext)


```

```python
%pylab inline
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
```

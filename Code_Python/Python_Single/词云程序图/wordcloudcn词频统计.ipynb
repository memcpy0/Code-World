```python
#encoding

with open("C:\\Users\\dell\\demo\\什么是心理学.txt",'r',encoding='utf-8') as f:
    mytext = f.read()
    
import jieba.analyse
result = jieba.analyse.textrank(mytext,topK=77,withWeight=True)#提取出词频高的词
```

```python
keywords = dict()
for i in result:
    keywords[i[0]]=i[1]
print(keywords)
```

```python
from PIL import Image,ImageSequence
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud,ImageColorGenerator
image = Image.open('4.jpg')
graph = np.array(image)
wc = WordCloud(font_path='./simsun.ttf',background_color='white',max_words=50,mask=graph)
wc.generate(mytext)
image_color = ImageColorGenerator(graph)
plt.imshow(wc)
plt.imshow(wc.recolor(color_func=image_color))
plt.axis("off")
plt.show()
```
